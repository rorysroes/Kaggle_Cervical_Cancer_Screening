```python
dim = 128
def showX(X, dim):
    int_X = X[:,:,:,::-1]
    int_X_reshape = int_X.reshape(-1,dim,dim,3).swapaxes(0,1).reshape(dim,-1,3)
    display(Image.fromarray(int_X_reshape))
#showX(train_data, dim)
```
